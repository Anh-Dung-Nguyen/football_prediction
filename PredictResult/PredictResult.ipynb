```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(3, closed = 'left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset = new_cols)
    return group
```

```python
cols = ["goals_for", "goals_against", "Possession %", "Shots on target", "Shots", "Touches", "Passes", "Tackles", "Clearances", "Corners", "Offsides", "Yellow cards", "Red cards"]
new_cols = [f"{c}_rolling" for c in cols]
```

```python
new_cols
```

```python
rolling_averages(group, cols, new_cols)
```

```python
matches_rolling = matches.groupby("team_name").apply(lambda x: rolling_averages(x, cols, new_cols))
```

```python
matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel('team_name')
```

```python
matches_rolling.index = range(matches_rolling.shape[0])
```

```python
def make_predictions(data, predictors):
    train = data[data["date"] < '2023-01-01']
    test = data[data["date"] >= '2023-01-01']
    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual = test["target"], predicted = preds), index = test.index)
    precision = precision_score(test["target"], preds, average = "macro")
    return combined, precision
```

```python
def predict_future_match(date_str, team_home, team_away):
    date = pd.to_datetime(date_str)
    day_code = date.dayofweek

    core_predictors = ["Side_code", "opp_code", "day_code"]
    all_predictors = core_predictors + new_cols

    def build_match_row(home_team, away_team):
        group = matches_rolling[matches_rolling["team_name"] == home_team]
        past_matches = group[group["date"] < date].sort_values("date").tail(3)

        if len(past_matches) < 3:
            raise ValueError(f"Not enough data for {home_team}")

        row = past_matches.mean(numeric_only = True)
        row["team_name"] = home_team
        row["opponent"] = away_team
        row["Side"] = "home"
        row["Side_code"] = 0

        if away_team in matches_rolling["opponent"].values:
            row["opp_code"] = matches_rolling[matches_rolling["opponent"] == away_team]["opp_code"].mode().values[0]
        else:
            row["opp_code"] = -1

        row["day_code"] = day_code
        row["date"] = date

        return row[all_predictors]

    match_row = build_match_row(team_home, team_away)
    match_df = pd.DataFrame([match_row])

    proba = rf.predict_proba(match_df)[0]

    print(f"Prediction from {team_home} (home) vs {team_away} (away)")
    print(f"{team_home} win: {proba[2]:.2%}")
    print(f"Draw: {proba[1]:.2%}")
    print(f"{team_away} win: {proba[0]:.2%}")

    pred = proba.argmax()
    if pred == 2:
        return f"Predict Winner: {team_home}"
    elif pred == 1:
        return f"Predict Winner: Draw"
    else:
        return f"Predict Winner: {team_away}"
```
